```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder
from  sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso #Гребневая регрессия
from sklearn.linear_model import Ridge #Гребневая регрессия
from sklearn.svm import SVR
from sklearn.metrics import log_loss
import sklearn.metrics
import pandas as pd
import numpy as np
import scipy
```

```python
import dask.dataframe as dd
import dask.array as da

df = pd.read_csv('CUP_IT_train_data.csv')
y = df['gi_smooth_3m']
df.drop(['gi_smooth_3m'], axis='columns', inplace=True)
X = df

df1 = pd.read_csv('CUP_IT_test_data.csv')
y_test = df1['gi_smooth_3m']
df1.drop(['gi_smooth_3m'], axis='columns', inplace=True)
X_test = df1

X = X.iloc[:, 9:]
X_test = X_test.iloc[:, 9:]
del X_test['rank']

for i in range(38):
    X.iloc[:, i].fillna(X.iloc[:, i].median(), inplace=True)
    X_test.iloc[:,i].fillna(X_test.iloc[:, i].median(), inplace=True)
```

### Пробую Деревья

```python
clf = RandomForestRegressor(n_estimators=34, random_state = 241)
clf.fit(X, y)
```

### Метрики для деревьев

```python
from sklearn.metrics import mean_absolute_error

yi = clf.predict(X_test)
yi = pd.Series(yi)

print(sklearn.metrics.mean_absolute_error(y_test, yi))

print(sklearn.metrics.mean_squared_error(y_test, yi))

print(sklearn.metrics.r2_score(y_test, yi))

print(sklearn.metrics.explained_variance_score(y_test, yi))

mape = mean_absolute_error(y_test, yi) / y_test.abs().sum()
print(mape * 100)
```

### Пробую Градиентный бустинг

```python
gb = GradientBoostingRegressor(n_estimators=30, random_state=0, learning_rate=0.2,  max_depth=2)
gb.fit(X, y)
```

### Метрики для градиентного бустинга

```python
from sklearn.metrics import mean_absolute_error

print(sklearn.metrics.mean_absolute_error(y_test, gb.predict(X_test)))
print(sklearn.metrics.r2_score(y_test, gb.predict(X_test)))

mape = mean_absolute_error(y_test, gb.predict(X_test)) / y_test.abs().sum()
print(mape * 100)
```

### Пробую Линейную регрессию

```python
clf = Lasso(alpha=0.1,random_state=241)
clf.fit(X, y)
```

### Метрики Лассо

```python
from sklearn.metrics import mean_absolute_error

yi = clf.predict(X_test)
yi = pd.Series(yi)

print(sklearn.metrics.mean_absolute_error(y_test, yi))

print(sklearn.metrics.mean_squared_error(y_test, yi))

print(sklearn.metrics.r2_score(y_test, yi))

print(sklearn.metrics.explained_variance_score(y_test, yi))

mape = mean_absolute_error(y_test, yi) / y_test.abs().sum()
print(mape * 100)

```
