```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import Lasso #Гребневая регрессия
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
import sklearn.metrics
import pandas as pd
import numpy as np
import scipy
```

```python
import dask.dataframe as dd
import dask.array as da

df = pd.read_csv('CUP_IT_train_data.csv')
y = df['gi_smooth_3m']
df.drop(['gi_smooth_3m'], axis='columns', inplace=True)
X = df

df1 = pd.read_csv('CUP_IT_test_data.csv')
y_test = df1['gi_smooth_3m']
df1.drop(['gi_smooth_3m'], axis='columns', inplace=True)
X_test = df1
```

```python
X.fillna('nan', inplace=True)
y.fillna('nan', inplace=True)
X_test.fillna('nan', inplace=True)
y_test.fillna('nan', inplace=True)

enc = DictVectorizer()
X_train = enc.fit_transform(X.to_dict('records')) #to_dict преобразовывает в словарь


clf = Lasso(alpha=0.1,random_state=241)
clf.fit(X_train, y)
```

```python
from sklearn.metrics import mean_absolute_error
X_test = enc.transform(X_test.to_dict('records'))
#X_test = enc1.transform(X_test)

yi = clf.predict(X_test)
yi = pd.Series(yi)

print(sklearn.metrics.mean_absolute_error(y_test, yi))

print(sklearn.metrics.mean_squared_error(y_test, yi))

print(sklearn.metrics.r2_score(y_test, yi))

print(sklearn.metrics.explained_variance_score(y_test, yi))

mape = mean_absolute_error(y_test, yi) / y_test.abs().sum()
print(mape * 100)
```
